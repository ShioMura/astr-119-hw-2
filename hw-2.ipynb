```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,2*np.pi,0.1)
plt.xlim(0,2*np.pi)
plt.ylim(-1,10)
y = 5.5*np.cos(2*x) + 5.5
plt.plot(x,y)
plt.xlabel('Time in ASTR119')
plt.ylabel('Measures of Awesomeness')
```

```python
x = np.arange(0,2*np.pi,0.1)
y = 0.02*np.exp(x)
plt.xlim(0,2*np.pi)
plt.ylim(-1,10)
plt.plot(x,y)
plt.xlabel('Time in ASTR119')
plt.ylabel('Measures of Awesomeness')
```

```python
x = np.arange(0,2*np.pi,0.1)
plt.xlim(0,2*np.pi)
plt.ylim(-1,10)
y = 0.25*x**2 + 0.1*np.sin(10*x)
plt.plot(x,y)
plt.xlabel('Time in ASTR119')
plt.ylabel('Measures of Awesomeness')
```
